```python
import requests # to download html code
from bs4 import BeautifulSoup # to navigate through the html code
import pandas as pd
import numpy as np
import re
```

```python
def scrape_hot100():
    url = 'https://www.billboard.com/charts/hot-100.'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    if response.status_code != 200:
        print("! Failed to download page content")
        return None

    songs   = []
    singers = []

    for x in soup.select('div.o-chart-results-list-row-container'):
        li = x.select('ul > li')[3]
        songs.append(li.h3.get_text().strip())
        singers.append(li.span.get_text().strip())
    
    df = pd.DataFrame({'Singer': singers,
                       'Song' : songs})
    #export to a csv file
    df.to_csv('hot_100.csv', index=False)

    return df
```

```python
hot100_df = scrape_hot100()
hot100_df
```
